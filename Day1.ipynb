#####**2. swap two varibales**


```python
#using temp variable
a = 10
b = 20
print("before swapping : ",end="")
print(f"a = {a}, b = {b}")

temp = a
a = b
b = temp
print("after swapping using temp : ",end="")
print(f"a = {a}, b = {b}")

#without using temp variable
#python style of swapping
a,b = b,a
print("after swapping using python style : ",end="")
print(f"a = {a}, b = {b}")

#swapping using logic
a = a+b
b = a-b
a = a-b
print("after swapping using Arithmetic logic : ",end="")
print(f"a = {a}, b = {b}")
```

#####**3. check given number is even or odd**

```python
num = int(input("Enter a number : "))
if num % 2 == 0:
  print(f'{num} is an even number')
else:
  print(f'{num} is an odd number')
```

#####**4. find the factorial**

```python
#Basic Programm
num = int(input("Enter the number : "))
fact_num = 1
if num < 0:
  print("Factorial does not exist for negative numbers")
elif num == 0:
  print("The factorial of 0 is 1")
else:
  for i in range(1,num+1):
    fact_num *= i
  print(f"The factorial of {num} is {fact_num}")
```

```python
#using Optimized function

def factorial(n: int) -> int:
  #define a local varibale
  _result = 1
  if n < 0:
    raise ValueError("Factorial does not exist for negative numbers")
  else:
    for i in range(2,n+1):
      _result *= i
    return _result

try:
  user_input = int(input("Enter a Positive Number : "))
  fact_number = factorial(user_input)
  print(f'The factorial of {user_input} is : {fact_number}')

except ValueError as e:
  print(e)
```

```python
#factorial using recursion

def factorial(n: int)->int:
  if n < 0:
    raise ValueError("Factorial is not defined for negative numbers")
  if n == 0 or n == 1:
    return 1
  return n*factorial(n-1)

try:
  user_input = int(input("Enter a Positive Number : "))
  fact_number = factorial(user_input)
  print(f'The factorial of {user_input} is : {fact_number}')

except ValueError as e:
  print(e)
```
