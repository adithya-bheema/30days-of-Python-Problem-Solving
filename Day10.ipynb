#####**Factorial using recursion**

```python
def factorial(n: int) -> int:
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers")

    if n == 0 or n == 1:
        return 1

    return n * factorial(n - 1)

try:
    user_input = int(input('enter the number : '))
    print(f'factorial of {user_input} is {factorial(user_input)}')
except ValueError as e:
    print(e)
```

#####**Fibbonacci series using recursion**

```python
#Basic version
def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)

num = int(input('enter the sequence length'))

for i in range(num):
    print(fib(i),end = ' ')
```

```python
#optimized by using memoization
def fibonacci(n: int, memo=None):
    if memo is None:
        memo = {0: 0, 1: 1}

    if n in memo:
        return memo[n]

    memo[n] = fibonacci(n - 1, memo) + fibonacci(n - 2, memo)
    return memo[n]


limit = int(input("Enter number of terms: "))

if limit <= 0:
    print("Enter a positive number")
else:
    sequence = []
    for i in range(limit):
        sequence.append(fibonacci(i))
    print(sequence)
```

#####**Prime factors of a number**

```python
num = int(input('enter the number : '))

prime_factors = []

i = 2

while num > 1:

    if num%i == 0:
        prime_factors.append(i)
        num //= i
    else:
        i += 1
print(f'prime factors of {num} are {prime_factors}')
```

```python
num = int(input("Enter the number: "))

prime_factors = []

# Step 1: Handle factor 2
while num % 2 == 0:
    prime_factors.append(2)
    num //= 2

# Step 2: Check only odd numbers up to sqrt(num)
i = 3
while i * i <= num:
    if num % i == 0:
        prime_factors.append(i)
        num //= i
    else:
        i += 2   # skip even numbers

# Step 3: If remaining number is prime
if num > 1:
    prime_factors.append(num)

print(f'prime factors of {num} are {prime_factors}')
```
