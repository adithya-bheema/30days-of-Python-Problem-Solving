#####**Buble sort**

```python
arr = [12,3,4,23,6,7,11,9,8,1,0]
print(f'original array : {arr}')

for i in range(len(arr)):
  for j in range(len(arr)-i-1):
    if arr[j] > arr[j+1]:
      arr[j],arr[j+1]=arr[j+1],arr[j]

print(f'sorted array : {arr}')
```

#####**Insertion sort**

```python
arr = [12,3,4,23,6,7,11,9,8,1,0]
print(f'original array : {arr}')

for i in range(1,len(arr)):
  key = arr[i]
  j = i - 1

  while j >= 0 and key < arr[j]:
    arr[j+1] = arr[j]
    j -= 1
  arr[j+1] = key

print(f'sorted array : {arr}')
```

#####**Linear search**

```python
arr = [12,3,4,23,6,7,11,9,8,1,0]

target = 6

for i in range(len(arr)):
  if arr[i] == target:
    print(f'element {target} found at index {i}')
    break

else:
    print('element not found!')
```

#####**Binary search**

```python
arr = [12,3,4,23,6,7,11,9,8,1,0]

#array should be sorted first
arr.sort()

target = 6
min = 0
max = len(arr) - 1

while min <= max:
  mid = min+max//2
  if target == arr[mid]:
    print(f'element {target} found at index {arr.index(target)}')
    break
  elif target > arr[mid]:
    min = mid
  else:
    max = mid
else:
  print('element not found')
```
