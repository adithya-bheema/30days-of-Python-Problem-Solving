#####**Remove duplicates from string**

```python
string = 'programming'

result = ''.join(sorted(set(string),key=string.index))
print(result)
```

```python
#best and optimized
string = 'programming'

result = ''.join(dict.fromkeys(string))
print(result)
```

```python
#manual logic
def remove_duplicates(s: str)->str:
  seen = set()
  unique = []
  for ch in s:
    if ch not in seen:
      seen.add(ch)
      unique.append(ch)
  return ''.join(unique)

user_string = 'programming'
print(f'string after duplicate removed : {remove_duplicates(user_string)}')
```

#####**Frequency counter**

```python
string = 'abbcdaabc'
freq = {}

for ch in string:
  freq[ch] = freq.get(ch,0)+1

print(freq)
```

```python
#without get method
string = 'abbcdaabc'
freq = {}

for ch in string:
    if ch not in freq:
        freq[ch] = 1
    else:
        freq[ch] += 1

print(freq)
```

#####**Reverse the string**

```python
#basic logic - not efficient
string = 'programming'
reverse = ''

for ch in string:
  reverse = ch + reverse

print(f'original string : {string}')
print(f'reversed string : {reverse}')
```

```python
#using string slicing - most efficient
string = 'programming'
reverse = string[::-1]

print(f'original string : {string}')
print(f'reversed string : {reverse}')
```

```python
#without slicing - not much efficient as slicing
string = 'programming'
chars = []

for ch in string:
    chars.insert(0, ch)

reverse = ''.join(chars)

print(f'original string : {string}')
print(f'reversed string : {reverse}')
```
