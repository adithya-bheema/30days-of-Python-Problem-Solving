#####**Pangram check**

```python
#using string module

import string

s = 'The quick brown fox jumps over the lazy dog'

result = set(string.ascii_lowercase) <= set(s.lower())

if result:
    print('The string is pangram!')
else:
    print('The given string is not a pangram')
```

```python
#Logic

def is_pangram(s:str)->bool:

    s = s.lower()

    for ch in range(ord('a'),ord('z')):
        if chr(ch) not in s:
            return False
    return True


string = 'The quick brown fox jumps over the lazy dog'

result = is_pangram(string)

if result:
    print('The string is pangram!')
else:
    print('The given string is not a pangram')
```

#####**Find unique elements in the list**

```python
my_list = [1,2,3,1,4,2,4,5,6,7,3,4]

unique = [x for x in my_list if my_list.count(x) == 1]

print(unique)
```

```python
my_list = [1,2,3,1,4,2,4,5,6,7,3,4]

freq = {}
for x in my_list:
    freq[x] = freq.get(x, 0) + 1

unique = [x for x in my_list if freq[x] == 1]

print(unique)
```

#####**Merge Two Stored Lists**

```python
#simple concatination
a = [1,3,4,7]
b = [0,6,5,2]

result = sorted(a+b)
print(result)
```

```python
#merge algorithm

a = [1,3,4,7]
b = [0,6,5,2]

a.sort()
b.sort()

i = j = 0
result = []

while i < len(a) and j < len(b):
    if a[i] < b[j]:
        result.append(a[i])
        i += 1
    else:
        result.append(b[j])
        j += 1

result.extend(a[i:])
result.extend(b[j:])

print(f'merged sorted list : {result}')
```
