#####sum of first n Natural numbers

```python
#Basic Program
num = int(input("Enter a number : "))
total = num * (num+1) // 2
print(f"The sum of first {num} natural numbers is {total}")
```

```python
#using loop

num = int(input("enter a number : "))
total = 0
for i in range(1,num+1):
  total += i
print(f"sum of first {num} natural numbers is : {total}")
```

```python
#optimized program
def nsum(num: int) -> int:
   if num <= 0:
    raise ValueError("Please enter a positive integer greater than 0")

   return num * (num+1) // 2

try:
  user_input = input("Enter a number : ")

  if not user_input.strip().isdigit():
    raise ValueError("Invalid input. Please enter a valid positive number.")

  num = int(user_input)
  total = nsum(num)
  print(f"The sum of first {num} natural numbers is {total}")
except ValueError as e:
  print(e)
```

#####Reverse the given string

```python
# reverse using string sclicing

user_string = input("Enter a string : ")
reversed_string = user_string[::-1]
print(f"The reversed string is : {reversed_string}")
```

```python
#reverse using loop
user_string = input("Enter a string : ")
reversed_string = ''
for char in user_string:
  reversed_string = char + reversed_string
print(f"The reversed string is : {reversed_string}")
```

```python
#optimized function

def reverse_string(s: str) -> str:
  if not isinstance(s, str):
    raise ValueError("Input must be a string")
  return s[::-1]

try:
  user_string = input("enter a string : ").strip()

  if not user_string:
    raise ValueError("Input string cannot be empty")

  reversed_string = reverse_string(user_string)
  print(f"The reversed string is : {reversed_string}")

except ValueError as e:
  print(e)

```

#####Palindrome Check

```python
#Basic program

user_string = input("Enter a string : ").strip()
reversed_string = user_string[::-1]

if user_string == reversed_string:
    print(f"The string {user_string} is a palindrome")
else:
    print(f"The string {user_string} is not a palindrome")
```

```python
#optimized program

def palindrome_check(s: str)->bool:

   """Below line eleminates the special symbols, spaces, punctuations"""
   cleaned = ''.join(s.lower() for c in s if c.isalnum())
   return cleaned == cleaned[::-1]

user_string = input("enter a string : ").strip()

if palindrome_check(user_string):
   print(f"The string {user_string} is a palindrome")
else:
   print(f"The string {user_string} is not a palindrome")
```

```python
#two pointer technique (avoids extra space to store reversed string)

def palindrome_check(s: str)->bool:

    cleaned_string = ''.join(c.lower() for c in s if c.isalnum())
    left, right = 0, len(cleaned_string)-1
    while left < right:
        if cleaned_string[left] != cleaned_string[right]:
            return False
        left += 1
        right -= 1
    return True

user_string = input("Enter a string : ").strip()

if palindrome_check(user_string):
    print(f"The string {user_string} is a palindrome")
else:
    print(f"The string {user_string} is not a palindrome")
```
