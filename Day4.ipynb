#####**Count Vowel in Given Stirng**

```python
user_string = input("Enter the String : ").strip().lower()
VOWELS = {'a','e','i','o','u'}
count = 0
for ch in user_string:
    if ch in VOWELS:
        count += 1
print(f'There are {count} vowels in "{user_string}"')
```

#####**Check Leap Year**

```python
year = int(input("enter the year : "))
if (year % 4 == 0 and year % 100 != 0) or year % 400 == 0:
    print(f"{year} is a leap year")
else:
    print(f"{year} is not a leap year")
```

```python
#optimized funtion
def is_leap_year(year: int) -> bool:
    return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)


try:
    year = int(input("Enter the year: ").strip())
    if is_leap_year(year):
        print(f"{year} is a leap year")
    else:
        print(f"{year} is not a leap year")
except ValueError:
    print("Please enter a valid year.")
```

#####**Celcius to Fharenheit**

```python
celsius = float(input("Enter temperature in Celsius: ").strip())
fahrenheit = (celsius * 9 / 5) + 32
print(f"{celsius}°C is equal to {fahrenheit:.2f}°F")
```

```python
#optimized function
def celsius_to_fahrenheit(celsius: float) -> float:
    return (celsius * 9 / 5) + 32

try:
    celsius = float(input("Enter temperature in Celsius: ").strip())
    fahrenheit = celsius_to_fahrenheit(celsius)
    print(f"{celsius}°C is equal to {fahrenheit:.2f}°F")
except ValueError:
    print("Please enter a valid temperature.")
```

#####**ASCII Value**

```python
char = input("Enter the character : ")
ascii = ord(char)
print(f'The ascii value of {char} is {ascii}')
```

```python
#optimized program
char = input("Enter a single ASCII character: ").strip()

if len(char) != 1:
    print("Please enter exactly one character.")
elif ord(char) > 127:
    print("Not an ASCII character.")
else:
    print(f"The ASCII value of '{char}' is {ord(char)}")
```

#####**Reverse a Number**

```python
num = int(input("Enter the Number : "))
rev = ''
while(num > 0):
    rev += str(num % 10)
    num //= 10

print(f"The reversed number : {rev}")
```

```python
#optimized program
try:
    num = int(input("Enter the number: ").strip())
    original = num

    rev = 0
    num = abs(num)

    while num > 0:
        rev = rev * 10 + (num % 10)
        num //= 10

    if original < 0:
        rev = -rev

    print(f"The reversed number is: {rev}")

except ValueError:
    print("Please enter a valid integer.")
```
