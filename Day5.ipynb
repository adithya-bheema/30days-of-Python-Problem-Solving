#####**Simple calculator**

```python
num1 = int(input("enter first number :"))
num2 = int(input("enter second number : "))
operation = input("enter operation to be performed : ")
match operation:
    case "addition":
        print(f"sum of {num1} and {num2} is {num1+num2}")
    case "subtraction":
        print(f"subtraction of {num1} and {num2} is {num1-num2}")
    case "multiplication":
        print(f"product {num1} and {num2} is {num1*num2}")
    case "division":
        print(f"division of {num1} and {num2} is {num1/num2}")
    case "power":
        print(f"{num1} exponent {num2} is {num1**num2}")
```

#####**Count occurrences of a character in string**

```python
user_input = input("enter the string : ").strip()
element = input("enter the element to find count : ").strip()

if len(element) != 1:
    print("enter only single character to find")
count = 0
for ch in user_input:
    if ch == element:
        count += 1
print(f"{element} occured {count} times")
```

#####**Find smallest element in a list**

```python
my_list = [12, 3, 4, 29, 34, 6, 7, 11, 2, 6, 9]

min_value = my_list[0]
for value in my_list[1:]:
    if value < min_value:
        min_value = value

print(f"The minimum element is: {min_value}")
```

#####**Check Armstrong number**

```python
num = input("Enter a number: ").strip()

if not num.isdigit():
    print("Please enter digits only.")
else:
    power = len(num)
    total = 0
    number = int(num)

    for ch in num:
        total += int(ch) ** power

    if total == number:
        print(f"{number} is an Armstrong number")
    else:
        print(f"{number} is not an Armstrong number")

```

```python
num = int(input("Enter a number: "))
original = num
power = len(str(num))
total = 0

while num > 0:
    digit = num % 10
    total += digit ** power
    num //= 10

if total == original:
    print(f"{original} is an Armstrong number")
else:
    print(f"{original} is not an Armstrong number")
```
