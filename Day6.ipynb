#####**Print all divisors of a number**

```python
#Basic
num = int(input("enter the number : "))
print(f'The divisors of {num} are : ',end='')
for i in range(1,int(num/2)+1):
    if num%i == 0:
        print(i,end=' ')
```

```python
#loop optimized
num = int(input("Enter the number: "))

print(f"The divisors of {num} are:", end=" ")
for i in range(1, int(num ** 0.5) + 1):
    if num % i == 0:
        print(i, end=" ")
        if i != num // i:
            print(num // i, end=" ")
```

```python
# fully optimized
try:
    num = int(input("Enter the number: ").strip())

    if num <= 0:
        print("Please enter a positive integer.")
    else:
        print(f"The divisors of {num} are:", end=" ")

        divisors = set()

        for i in range(1, int(num ** 0.5) + 1):
            if num % i == 0:
                divisors.add(i)
                divisors.add(num // i)

        for d in sorted(divisors):
            print(d, end=" ")

except ValueError:
    print("Enter a valid integer.")
```

#####**Find length of string without len()**

```python
user_string = input("Enter the string: ")

length = 0
for _ in user_string:
    length += 1

print(f'The length of "{user_string}" is {length}')
```

#####**Print ASCII characters from A to Z**

```python
print("ASCII values for A–Z:")
for i in range(65, 91):
    print(f"{i} -> {chr(i)}")
```

#####**Sum of elements in a list**

```python
my_list = [12,34,54,7,8,16,27]
total = 0
for i in my_list:
    total += i

print(f"The sum of elements of the list is : {total}")
```

#####**Remove duplicates from list**

```python
#Basic but no gurantee on order of elements

my_list = [1,2,7,3,2,3,4,1,5,6,7,4,6,3,2,1]
unique_list = list(set(my_list))

print(f'unique list : {unique_list}')
```

```python
#Basic and preserving order of elements

my_list = [1,2,7,3,2,3,4,1,5,6,7,4,6,3,2,1]
unique_list = list(dict.fromkeys(my_list))

print(f'unique list : {unique_list}')
```

```python
my_list = [1,2,3,2,3,4,1,5,6,7,4,6,3,2,1]
unique_list = []

for item in my_list:
    if item not in unique_list:
        unique_list.append(item)

print(f'unique list : {unique_list}')
```
