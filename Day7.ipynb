#####**Convert Decimal to Binary**

```python
#with built-in
num = 20
bin_num = bin(num)
print(f'Binary number of {num} is {bin_num[2:]}')
```

```python
#without builtin
num = 38
bin_num = ''

while num > 0:
    bin_num = str(num%2) + bin_num
    num //=2
print(f'Binary number of {num} is {bin_num[2:]}')
```

#####**Convert Binary to Decimal**

```python
#Basic Conversion
bin_num = "1110"
num = int(bin_num,2)
print(f'Decimal number of {bin_num} is {num}')
```

```python
#logical method
bin_num = "10100"
num = 0

for i in range(len(bin_num)):
    if bin_num[-(i+1)] == '1':
        num += 2 ** i

print(f'Decimal number of {bin_num} is {num}')

```
