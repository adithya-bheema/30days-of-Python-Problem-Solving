#####**Find GCD (Euclidean Algorithm)**

```python
#using built-in
import math

a,b = 21,6
print(f'GCD of {a} and {b} is {math.gcd(a,b)}')
```

```python
#using manual function
def find_gcd(a,b):
  while b != 0:
      a,b = b, a%b
  return a

num1 = int(input('enter the first number : '))
num2 = int(input('enter the second number : '))
print(f'The GCD of {num1} and {num2} is : {find_gcd(num1,num2)}')
```

#####**Find LCM of two numbers**

```python
#using built-in
import math
a = 10
b = 15
result = abs(a*b)//math.gcd(a,b)
print(f'LCM of given two {a} and {b} is {result}')
```

```python
#using manual function
def find_lcm(a,b):
    greater = max(a,b)

    while True:
        if greater % b == 0 and greater % a == 0:
            return greater
        greater += 1


num1 = int(input('enter the first number : '))
num2 = int(input('enter the second number : '))
result = find_lcm(num1,num2)
print(f'LCM of given two {num1} and {num2} is {result}')
```

#####**Armstrong number**

```python
#Basic program
num = input("enter the number: ")
power = len(num)

result = sum(int(i) ** power for i in num)

if result == int(num):
    print("It is an Armstrong number")
else:
    print("It is not an Armstrong number")
```

```python
#with input validation
num = input("Enter a number: ").strip()

if not num.isdigit():
    print("Please enter digits only.")
else:
    power = len(num)
    number = int(num)

    result = sum(int(d)**power for d in num)

    if result == number:
        print(f"{number} is an Armstrong number")
    else:
        print(f"{number} is not an Armstrong number")
```

#####**Count words in a string**

```python
#builtin split function
string = input('write any string :')
word_count = len(string.split())
print(f"Total words : {word_count}")
```

```python
#using logic
string = input("Write any string: ").strip()

count = 0
in_word = False

for ch in string:
    if ch != " ":
       if not in_word:
            count += 1
            in_word = True
    else:
        in_word = False

print(f"Total words: {count}")
```

#####**Check if two strings are anagrams**

```python
word1 = input('enter any word : ')
word2 = input('enter another word : ')
if sorted(word1) == sorted(word2):
    print('Given words are anagrams!')
else:
    print('Given words are not anagrams')
```

```python
word1 = input('enter any word : ')
word2 = input('enter another word : ')

if len(word1) != len(word2):
    print('Given words are not anagrams..')
else:
    freq = {}
    for ch in word1:
        freq[ch] = freq.get(ch,0) + 1
    for ch in word2:
        if ch not in freq or freq[ch] == 0:
            print('Given words are not anagrams..')
            break
        freq[ch] -= 1
    else:
        print('Given words are anagrams!')
```
