#####**Unique elements in a List**

```python
#using built-in logic

my_list = [12,34,5,5,6,11,10,34,11,25,12,9]

unique_list = list(set(my_list))

print(f'unique elements are : {unique_list}')
```

```python
#using manual inspection

my_list = [12,34,5,5,6,11,10,34,11,25,12,9]

unique_list = []

for num in my_list:
    found = False
    for ele in unique_list:
        if ele == num:
            found = True
            break

    if not found:
        unique_list.append(num)

print(f'unique elements are : {unique_list}')
```

#####**Second Largest element in a List**

```python
#using built-in

my_list = [12,34,5,5,6,11,10,34,11,25,12,9]

result = sorted(set(my_list))[-2]
print(f'Second Largest Number is : {result}')
```

```python
#optimized version

my_list = [12,34,5,5,6,11,10,34,11,25,12,9]

unique = set(my_list)
first = second = float('-inf') #initialzing to -infinity(lowest value)

for num in unique:
    if num > first:
        second = first
        first = num
    elif first > num > second:
        second = num

print(f'Second Largest Number is : {second}')
```

#####**Transpose of a Matrix**

```python
#using built-in

mat = [[1,2,3],[4,5,6]]
trp = [list(row) for row in (zip(*mat))]
print(f'Transpose of matrix {mat} is : {trp}')
```

```python
#Logical way
mat = [[1,2,3],[4,5,6]]

#initialize rows and cols for transpose matrix
rows = len(mat)
cols = len(mat[0])

#initalizing transepose matrix with zeros
trp = [[0]*rows for _ in range(cols)]

for r in range(rows):
    for c in range(cols):
        trp[c][r] = mat[r][c]

print(f'Transpose of matrix {mat} is : {trp}')
```
