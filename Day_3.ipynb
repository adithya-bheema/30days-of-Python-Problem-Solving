#####**Fibbonacci Series**

```python
#basic program
limit = int(input("Enter no. of elements: "))
first = 0
second = 1
for i in range(limit):
    print (first, end = " ")
    first,second = second,first+second
```

```python
#optimized function

def fibbonacci(n: int)->list:
    if n <= 0:
       raise ValueError("Enter a POSITIVE integer!")

    '''Here we are not using append() function.
     so We initialzing a list with Zeros of length n'''

    result = [0]*n
    first, second = 0, 1

    for i in range(n):
        result[i] = first
        first, second = second, first + second

    return result

try:
    limit = int(input("enter required no. of sequence : "))
    sequence = fibbonacci(limit)
    print(f"sequence: {sequence}")

except ValueError as e:
    print(e)
```

#####**Max of Three Numbers**

```python
#basic program with built-in function max()

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
num3 = int(input("Enter third number: "))
print(f"Maximum of three numbers is {max(num1,num2,num3)}")
```

```python
#Basic Comparision logic Program
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
num3 = int(input("Enter third number: "))

print("The maximum number is : ", end = "")
if num1 >= num2 and num1 >= num3:
    print(num1)
elif num2 >= num1 and num2 >= num3:
    print(num2)
else:
    print(num3)
```

```python
#optimized function

def max_of_three(num1:int,num2:int,num3:int) -> int:

    max_num = num1

    if num2 > max_num:
        max_num = num2

    if num3 > max_num:
        max_num = num3

    return max_num

try:
    first = int(input("Enter first number: "))
    second = int(input("Enter second number: "))
    third = int(input("Enter third number: "))

    result = max_of_three(first,second,third)
    print(f"Maximum of three numbers is {result}")

except ValueError:
    print("Enter a valid integer!")
```

#####**Prime Number**

```python
# basic Program

num = int(input("Enter a number: "))
for i in range(2,num):
    if num % i == 0:
        print(f"{num} is not a prime number")
        break
else:
    print(f"{num} is a prime number")
```

```python
#optimized program

def prime_check(num:int)->bool:

    if num < 2:
        return False

    for i in range(2,int(num**0.5)+1):
        if num % i == 0:
            return False

    return True


try:
    user_input = int(input("Enter the number : "))

    if user_input <= 0:
        raise ValueError()

    if prime_check(user_input):
        print(f"{user_input} is a prime number")
    else:
        print(f"{user_input} is not a prime number")

except ValueError:
    print("Enter a valid POSITIVE integer!")
```

#####**Sum of Digits**

```python
#basic program
num = input("Enter the number : ")

total = 0
for digit in num:
    total += int(digit)
print(total)
```

```python
#optimized program using buitin validation
num = input("Enter the number : ")

total = 0
for digit in num:
    if digit.isdigit():
        total += int(digit)

print(f"The sum of digits in number {num} is : {total}")

```

```python
#optimized program using mathematical approach
user_num = input("Enter the number : ").strip()
total = 0
try:
    num = abs(int(user_num))
    while num > 0:
        total += num % 10
        num //= 10

    print(f"The sum of digits in number {user_num} is : {total}")
except ValueError:
    print("Enter a valid integer Number")
```

#####**Largest number in the List**

```python
#basic using loop

l1 = [10,4,3,20,5,130]
max_num = l1[0]
for i in range(1,len(l1)):
    if max_num < l1[i]:
        max_num = l1[i]
print(f"The maximum element is : {max_num}")
```

#####**Multiplication table of a number**

```python
#basic
num = int(input("Enter which table you need : "))
print(f"The {num} table is : ")
for i in range(1,11):
    res = num * i
    print(f"{num} x {i} = {res}")
```

```python
#optimized

try:
    num = int(input("Enter which table you need : ").strip())
    limit = int(input("Print up to which number: ").strip())
    print(f"Multiplication table of {num} upto {limit} :")

    for i in range(1,limit+1):
        print(f"{num} x {i} = {num*i}")
except ValueError:
    print("Enter valid numbers!")
```
